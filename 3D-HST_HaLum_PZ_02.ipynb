```python
#### Function to calculate the fraction of area in a given redshift range in the probability distribution of redshift

def pz_corr(specid,zint,zfin):
    filename=file1+specid+'.new_zfit.pz.fits'
    
    hdulist1=pyfits.open(filename)
    
    zpdf=hdulist1[4].data ### REDSHIFT
    pboth=hdulist1[5].data ### PROBABILITY (in log units)
    
    ### CALCULATING THE TOTAL AREA UNDER THE PDF
    
    diff=np.diff(zpdf) ## CALCULATING THE DX DIFFERENTIAL ELEMENT
    dx=diff.tolist()### MAKING IT TO A LIST
    dx.append(diff[-1]) ## SETTING BOUNDARY AT THE LAST ELEMENT
    tot_area=sum((10**pboth)*dx) ### CALCULATING THE AREA UNDER THE PDF
    
    ### CALCULATING THE PARTIAL AREA BELOW THE BEST CALCULATED REDSHIFT
    
    z_bel=[]
    pbel=[]
    
    for k in range(0,len(zpdf)):
        if zint<=zpdf[k]<=zfin:
            z_bel.append(zpdf[k])
            pbel.append(pboth[k])
        
    z_bel=np.array(z_bel)
    pbel=np.array(pbel)
    
    diff_bel=np.diff(z_bel)
    dx_bel=diff_bel.tolist()
    dx_bel.append(diff_bel[-1])
    part_area=sum((10**pbel)*dx_bel)
    frac= part_area/tot_area
    
    
    return frac
```

```python
######################## Calculating Luminosity from the Ha Line Flux #############################
#Determining the luminosity distance (in units of Mpc) from redshift using Astropy
dA_L=[]
dC_L=[]
dGN_L=[]
dGS_L=[]
dU_L=[]

for i in range(0,len(gA_zgm)):
    d_l=distfunc(gA_zgm[i])
    dA_L.append(d_l)
    
for i in range(0,len(gC_zgm)):
    d_l=distfunc(gC_zgm[i])
    dC_L.append(d_l)

for i in range(0,len(gGN_zgm)):
    d_l=distfunc(gGN_zgm[i])
    dGN_L.append(d_l)

for i in range(0,len(gGS_zgm)):
    d_l=distfunc(gGS_zgm[i])
    dGS_L.append(d_l)

for i in range(0,len(gU_zgm)):
    d_l=distfunc(gU_zgm[i])
    dU_L.append(d_l)
    

dA_L=np.array(dA_L) # in units of Mpc
dC_L=np.array(dC_L)
dGN_L=np.array(dGN_L)
dGS_L=np.array(dGS_L)
dU_L=np.array(dU_L)

dA_Lcm=dA_L*3.08567758*10**24 # Converting from Mpc to cm
dC_Lcm=dC_L*3.08567758*10**24
dGN_Lcm=dGN_L*3.08567758*10**24
dGS_Lcm=dGS_L*3.08567758*10**24
dU_Lcm=dU_L*3.08567758*10**24
```

```python
######################## Flux to Luminosity Conversion #############################
# Calculating the luminosity from the distance and the flux

LumA=[]
LumC=[]
LumGN=[]
LumGS=[]
LumU=[]

fA_CC=[]
fC_CC=[]
fGN_CC=[]
fGS_CC=[]
fU_CC=[]

gA_z=[]
gC_z=[]
gGN_z=[]
gGS_z=[]
gU_z=[]

for i in range(0,len(gA_haflux)):
    if gA_haflux[i]>=3:
        L=4*np.pi*dA_Lcm[i]*dA_Lcm[i]*gA_haflux[i]*10**-17
        gA_z.append(gA_zgm[i])
        LumA.append(L)
        fA_CC.append(CCA[i])
gA_z=np.array(gA_z)
LumA=np.array(LumA)
fA_CC=np.array(fA_CC)


for i in range(0,len(gC_haflux)):
    if gC_haflux[i]>=3:
        L=4*np.pi*dC_Lcm[i]*dC_Lcm[i]*gC_haflux[i]*10**-17
        gC_z.append(gC_zgm[i])
        LumC.append(L)
        fC_CC.append(CCC[i])
gC_z=np.array(gC_z)
LumC=np.array(LumC)
fC_CC=np.array(fC_CC)

for i in range(0,len(gGN_haflux)):
    if gGN_haflux[i]>=3:
        L=4*np.pi*dGN_Lcm[i]*dGN_Lcm[i]*gGN_haflux[i]*10**-17
        gGN_z.append(gGN_zgm[i])
        LumGN.append(L)
        fGN_CC.append(CCGN[i])
gGN_z=np.array(gGN_z)
LumGN=np.array(LumGN)
fGN_CC=np.array(fGN_CC)

for i in range(0,len(gGS_haflux)):
    if gGS_haflux[i]>=3:
        L=4*np.pi*dGS_Lcm[i]*dGS_Lcm[i]*gGS_haflux[i]*10**-17
        gGS_z.append(gGS_zgm[i])
        LumGS.append(L)
        fGS_CC.append(CCGS[i])
gGS_z=np.array(gGS_z)
LumGS=np.array(LumGS)
fGS_CC=np.array(fGS_CC)

for i in range(0,len(gU_haflux)):
    if gU_haflux[i]>=3:
        L=4*np.pi*dU_Lcm[i]*dU_Lcm[i]*gU_haflux[i]*10**-17
        gU_z.append(gU_zgm[i])
        LumU.append(L)
        fU_CC.append(CCU[i])
gU_z=np.array(gU_z)
LumU=np.array(LumU)
fU_CC=np.array(fU_CC)
```

```python
######################## Calculating Vmax #############################

# The Flux Limit is:
flux_limit=3.0 # units of 10^-17 ergs/s/cm^2. We've already filter out the fluxes that are below
               # this limit.

zAmax=[]
dALmax=[]

zCmax=[]
dCLmax=[]

zGNmax=[]
dGNLmax=[]

zGSmax=[]
dGSLmax=[]

zUmax=[]
dULmax=[]

for i in range(0,len(LumA)):
    dl=math.sqrt(LumA[i]/(4*np.pi*flux_limit*10**-17))*(3.24077929*10**-25)
    z=redsfunc(dl)
    zAmax.append(z)
    dALmax.append(dl)
zAmax=np.array(zAmax)
dALmax=np.array(dALmax)

for i in range(0,len(LumC)):
    dl=math.sqrt(LumC[i]/(4*np.pi*flux_limit*10**-17))*(3.24077929*10**-25)
    z=redsfunc(dl)
    zCmax.append(z)
    dCLmax.append(dl)
zCmax=np.array(zCmax)
dCLmax=np.array(dCLmax)


for i in range(0,len(LumGN)):
    dl=math.sqrt(LumGN[i]/(4*np.pi*flux_limit*10**-17))*(3.24077929*10**-25)
    z=redsfunc(dl)
    zGNmax.append(z)
    dGNLmax.append(dl)
zGNmax=np.array(zGNmax)
dGNLmax=np.array(dGNLmax)

for i in range(0,len(LumGS)):
    dl=math.sqrt(LumGS[i]/(4*np.pi*flux_limit*10**-17))*(3.24077929*10**-25)
    z=redsfunc(dl)
    zGSmax.append(z)
    dGSLmax.append(dl)
zGSmax=np.array(zGSmax)
dGSLmax=np.array(dGSLmax)


for i in range(0,len(LumU)):
    dl=math.sqrt(LumU[i]/(4*np.pi*flux_limit*10**-17))*(3.24077929*10**-25)
    z=redsfunc(dl)
    zUmax.append(z)
    dULmax.append(dl)
zUmax=np.array(zUmax)
dULmax=np.array(dULmax)

### Calculating the comoving volume in Mpc^3
VmaxA=cd.comoving_volume(zAmax, **cosmo) # in units of Mpc^3  
VmaxC=cd.comoving_volume(zCmax, **cosmo) # in units of Mpc^3  
VmaxGN=cd.comoving_volume(zGNmax, **cosmo) # in units of Mpc^3 
VmaxGS=cd.comoving_volume(zGSmax, **cosmo) # in units of Mpc^3  
VmaxU=cd.comoving_volume(zUmax, **cosmo) # in units of Mpc^3  


Vmin1=cd.comoving_volume(0.7, **cosmo) # Minimum volume for redshift range 1
Vmin2=cd.comoving_volume(1.1, **cosmo) # Minimum volume for redshift range 2

Vmx1=cd.comoving_volume(1.1, **cosmo) # Maximum volume for redshift range 1
Vmx2=cd.comoving_volume(1.5, **cosmo) # Maximum volume for redshift range 2
```

```python
# Grouping the Haplha emitters into two redshift regimes (0.7-1.1 and 1.1 to 1.5)

A_z1=[]
A_L1=[]
A_Vmax1=[]
A_CC1=[]

A_z2=[]
A_L2=[]
A_Vmax2=[]
A_CC2=[]

for i in range(0,len(gA_z)):
    if 0.7<=gA_z[i]<1.1:
        A_z1.append(gA_z[i])
        A_L1.append(LumA[i])
        A_CC1.append(fA_CC[i])
        if VmaxA[i]<Vmx1:
            term=(VmaxA[i]-Vmin1)/(129600.0/np.pi)*A_area
        else:
            term=(Vmx1-Vmin1)/(129600.0/np.pi)*A_area
        A_Vmax1.append(term)
        
    elif 1.1<=gA_z[i]<1.5:
        A_z2.append(gA_z[i])
        A_L2.append(LumA[i])
        A_CC2.append(fA_CC[i])
        if VmaxA[i]<Vmx2:
            term=(VmaxA[i]-Vmin2)/(129600.0/np.pi)*A_area
        else:
            term=(Vmx2-Vmin2)/(129600.0/np.pi)*A_area
        A_Vmax2.append(term)        

        
C_z1=[]
C_L1=[]
C_Vmax1=[]
C_CC1=[]

C_z2=[]
C_L2=[]
C_Vmax2=[]
C_CC2=[]

for i in range(0,len(gC_z)):
    if 0.7<=gC_z[i]<1.1:
        C_z1.append(gC_z[i])
        C_L1.append(LumC[i])
        C_CC1.append(fC_CC[i])
        if VmaxC[i]<Vmx1:
            term=(VmaxC[i]-Vmin1)/(129600.0/np.pi)*C_area
        else:
            term=(Vmx1-Vmin1)/(129600.0/np.pi)*C_area
        C_Vmax1.append(term)
        
    elif 1.1<=gC_z[i]<1.5:
        C_z2.append(gC_z[i])
        C_L2.append(LumC[i])
        C_CC2.append(fC_CC[i])
        if VmaxC[i]<Vmx2:
            term=(VmaxC[i]-Vmin2)/(129600.0/np.pi)*C_area
        else:
            term=(Vmx2-Vmin2)/(129600.0/np.pi)*C_area
        C_Vmax2.append(term)        

GN_z1=[]
GN_L1=[]
GN_Vmax1=[]
GN_CC1=[]

GN_z2=[]
GN_L2=[]
GN_Vmax2=[]
GN_CC2=[]

for i in range(0,len(gGN_z)):
    if 0.7<=gGN_z[i]<1.1:
        GN_z1.append(gGN_z[i])
        GN_L1.append(LumGN[i])
        GN_CC1.append(fGN_CC[i])
        if VmaxGN[i]<Vmx1:
            term=(VmaxGN[i]-Vmin1)/(129600.0/np.pi)*GN_area
        else:
            term=(Vmx1-Vmin1)/(129600.0/np.pi)*GN_area
        GN_Vmax1.append(term)
        
    elif 1.1<=gGN_z[i]<1.5:
        GN_z2.append(gGN_z[i])
        GN_L2.append(LumGN[i])
        GN_CC2.append(fGN_CC[i])
        if VmaxGN[i]<Vmx2:
            term=(VmaxGN[i]-Vmin2)/(129600.0/np.pi)*GN_area
        else:
            term=(Vmx2-Vmin2)/(129600.0/np.pi)*GN_area
        GN_Vmax2.append(term)        

GS_z1=[]
GS_L1=[]
GS_Vmax1=[]
GS_CC1=[]

GS_z2=[]
GS_L2=[]
GS_Vmax2=[]
GS_CC2=[]

for i in range(0,len(gGS_z)):
    if 0.7<=gGS_z[i]<1.1:
        GS_z1.append(gGS_z[i])
        GS_L1.append(LumGS[i])
        GS_CC1.append(fGS_CC[i])
        if VmaxGS[i]<Vmx1:
            term=(VmaxGS[i]-Vmin1)/(129600.0/np.pi)*GS_area
        else:
            term=(Vmx1-Vmin1)/(129600.0/np.pi)*GS_area
        GS_Vmax1.append(term)
        
    elif 1.1<=gGS_z[i]<1.5:
        GS_z2.append(gGS_z[i])
        GS_L2.append(LumGS[i])
        GS_CC2.append(fGS_CC[i])
        if VmaxGS[i]<Vmx2:
            term=(VmaxGS[i]-Vmin2)/(129600.0/np.pi)*GS_area
        else:
            term=(Vmx2-Vmin2)/(129600.0/np.pi)*GS_area
        GS_Vmax2.append(term)        


U_z1=[]
U_L1=[]
U_Vmax1=[]
U_CC1=[]

U_z2=[]
U_L2=[]
U_Vmax2=[]
U_CC2=[]

for i in range(0,len(gU_z)):
    if 0.7<=gU_z[i]<1.1:
        U_z1.append(gU_z[i])
        U_L1.append(LumU[i])
        U_CC1.append(fU_CC[i])
        if VmaxU[i]<Vmx1:
            term=(VmaxU[i]-Vmin1)/(129600.0/np.pi)*U_area
        else:
            term=(Vmx1-Vmin1)/(129600.0/np.pi)*U_area
        U_Vmax1.append(term)
        
    elif 1.1<=gU_z[i]<1.5:
        U_z2.append(gU_z[i])
        U_L2.append(LumU[i])
        U_CC2.append(fU_CC[i])
        if VmaxU[i]<Vmx2:
            term=(VmaxU[i]-Vmin2)/(129600.0/np.pi)*U_area
        else:
            term=(Vmx2-Vmin2)/(129600.0/np.pi)*U_area
        U_Vmax2.append(term)        


A_z1=np.array(A_z1)
A_L1=np.array(A_L1)
A_Vmax1=np.array(A_Vmax1)
A_CC1=np.array(A_CC1)

A_z2=np.array(A_z2)
A_L2=np.array(A_L2)
A_Vmax2=np.array(A_Vmax2)
A_CC2=np.array(A_CC2)

C_z1=np.array(C_z1)
C_L1=np.array(C_L1)
C_Vmax1=np.array(C_Vmax1)
C_CC1=np.array(C_CC1)

C_z2=np.array(C_z2)
C_L2=np.array(C_L2)
C_Vmax2=np.array(C_Vmax2)
C_CC2=np.array(C_CC2)

GN_z1=np.array(GN_z1)
GN_L1=np.array(GN_L1)
GN_Vmax1=np.array(GN_Vmax1)
GN_CC1=np.array(GN_CC1)

GN_z2=np.array(GN_z2)
GN_L2=np.array(GN_L2)
GN_Vmax2=np.array(GN_Vmax2)
GN_CC2=np.array(GN_CC2)

GS_z1=np.array(GS_z1)
GS_L1=np.array(GS_L1)
GS_Vmax1=np.array(GS_Vmax1)
GS_CC1=np.array(GS_CC1)

GS_z2=np.array(GS_z2)
GS_L2=np.array(GS_L2)
GS_Vmax2=np.array(GS_Vmax2)
GS_CC2=np.array(GS_CC2)

U_z1=np.array(U_z1)
U_L1=np.array(U_L1)
U_Vmax1=np.array(U_Vmax1)
U_CC1=np.array(U_CC1)

U_z2=np.array(U_z2)
U_L2=np.array(U_L2)
U_Vmax2=np.array(U_Vmax2)
U_CC2=np.array(U_CC2)

# Applying correction to 1/Vmax

A_CCVmax1=(1/(A_Vmax1*A_CC1))
A_CCVmax2=(1/(A_Vmax2*A_CC2))

C_CCVmax1=(1/(C_Vmax1*C_CC1))
C_CCVmax2=(1/(C_Vmax2*C_CC2))

GN_CCVmax1=(1/(GN_Vmax1*GN_CC1))
GN_CCVmax2=(1/(GN_Vmax2*GN_CC2))

GS_CCVmax1=(1/(GS_Vmax1*GS_CC1))
GS_CCVmax2=(1/(GS_Vmax2*GS_CC2))

U_CCVmax1=(1/(U_Vmax1*U_CC1))
U_CCVmax2=(1/(U_Vmax2*U_CC2))
```
