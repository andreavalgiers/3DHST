```python
######################## Cosmology Parameters, Luminosity Distance, and Redshift Function #############################
### We have assumed the following cosmological parameters:
# Omega_m=0.3, Omega_L=0.7, Omega_k=0, h=0.70

cosmo = {'omega_M_0' : 0.3, 'omega_lambda_0' : 0.7, 'h' : 0.70}
cosmo = cd.set_omega_k_0(cosmo)
distfunc, redsfunc = cd.quick_distance_function(cd.luminosity_distance, return_inverse=True, **cosmo)

# Check for z=6.3333
d = distfunc(6.3333) # Calculates luminosity distance as a function of redshift
z = redsfunc(d) # Calculates redshift as a function of luminosity distance
print d
print z
```

```python
######################## Calculating Luminosity from the Ha Line Flux #############################
#Determining the luminosity distance (in units of Mpc) from redshift using Astropy
d_L=[]
for i in range(0,len(g_zgm)):
    d_l=distfunc(g_zgm[i])
    d_L.append(d_l)
    

d_L=np.array(d_L) # in units of Mpc

d_Lcm=d_L*3.08567758*10**24 # Converting from Mpc to cm

# Checking conversion from Mpc to cm
print g_zgm[0]
print d_L[0]
print d_Lcm[0]
```

```python
######################## Flux to Luminosity Conversion #############################
# Calculating the luminosity from the distance and the flux

Lum=[]
f_CC=[]
g_z=[]
gf140mag=[]
for i in range(0,len(g_oflux)):
    if g_oflux[i]>=3: # Here we are applying the flux limit
        L=4*np.pi*d_Lcm[i]*d_Lcm[i]*g_oflux[i]*10**-17
        g_z.append(g_zgm[i])
        Lum.append(L)
        f_CC.append(CC[i])
        gf140mag.append(g_F140mag[i])
g_z=np.array(g_z)
Lum=np.array(Lum)
f_CC=np.array(f_CC)
gf140mag=np.array(gf140mag)
```

```python
######################## Calculating Volume for Flux Limited #############################

# The Flux Limit is:
flux_limit=3.0 # units of 10^-17 ergs/s/cm^2. We've already filter out the fluxes that are below
               # this limit.

zmax=[]
dLmax=[]

for i in range(0,len(Lum)):
    dl=math.sqrt(Lum[i]/(4*np.pi*flux_limit*10**-17))*(3.24077929*10**-25) # In units of Mpc
    z=redsfunc(dl)
    zmax.append(z)
    dLmax.append(dl)
zmax=np.array(zmax)
dLmax=np.array(dLmax)

### Calculating the comoving volume in Mpc^3
V=cd.comoving_volume(zmax, **cosmo) # in units of Mpc^3  
```

```python
# Grouping the Haplha emitters into two redshift regimes (0.7-1.1 and 1.1 to 1.5)

z1=[]
L1=[]
V1=[]
CC1=[]
f140W1=[]

z2=[]
L2=[]
V2=[]
CC2=[]
f140W2=[]

Vmin1=cd.comoving_volume(1.0, **cosmo)
Vmin2=cd.comoving_volume(1.5, **cosmo)

Vmx1=cd.comoving_volume(1.5, **cosmo)
Vmx2=cd.comoving_volume(2.3, **cosmo)


for i in range(0,len(V)):
    if 1.0<=g_z[i]<1.5:
        z1.append(g_z[i])
        L1.append(Lum[i])
        CC1.append(f_CC[i])
        f140W1.append(gf140mag[i])
        if V[i]>Vmx1:
            V1.append(Vmx1-Vmin1)
        else:
            V1.append(V[i]-Vmin1)
        
    elif 1.5<=g_z[i]<2.3:
        z2.append(g_z[i])
        L2.append(Lum[i])
        CC2.append(f_CC[i])
        f140W2.append(gf140mag[i])
        if V[i]>Vmx2:
            V2.append(Vmx2-Vmin2)
        else:
            V2.append(V[i]-Vmin2)
        
z1=np.array(z1)
L1=np.array(L1)
V1=np.array(V1)
CC1=np.array(CC1)
f1410W1=np.array(f140W1)

z2=np.array(z2)
L2=np.array(L2)
V2=np.array(V2)
CC2=np.array(CC2)
f1410W2=np.array(f140W2)
```

```python
######################## Correcting Vmax for Field of View #############################

Vmax1=(V1/(129600.0/np.pi))*(0.032535+0.033774+0.032132+0.040972+0.03209)
Vmax2=(V2/(129600.0/np.pi))*(0.032535+0.033774+0.032132+0.040972+0.03209)

# Applying correction to 1/Vmax

CCVmax1=(1/(Vmax1*CC1))
CCVmax2=(1/(Vmax2*CC2))
```

```python
######################## Binning the luminosity in each redshift regime #############################
### In the redshift regime: 1.0 < z < 1.5

n=0.25
binwidth01=n
bins01=np.arange(min(np.log10(L1)), max(np.log10(L1)) + binwidth01, binwidth01)

phi1,binr1,cint1=stats.binned_statistic(np.log10(L1),(1/Vmax1)*(1/CC1), statistic='sum', bins=bins01)
count1,binr1,cint1=stats.binned_statistic(np.log10(L1),(1/Vmax1)*(1/CC1), statistic='count', bins=bins01)
logPhi1=np.log10(phi1/binwidth01)

# Creating the bin:
logL1=[]
for i in range(0,len(binr1)-1):
    logL1.append((binwidth01/2)+binr1[i])
logL1=np.array(logL1)

### In the redshift regime: 1.5 < z < 2.3

binwidth02=n
bins02=np.arange(min(np.log10(L2)), max(np.log10(L2)) + binwidth02, binwidth02)

phi2,binr2,cint2=stats.binned_statistic(np.log10(L2),(1/Vmax2)*(1/CC2), statistic='sum', bins=bins02)
count2,binr2,cint2=stats.binned_statistic(np.log10(L2),(1/Vmax2)*(1/CC2), statistic='count', bins=bins02)
logPhi2=np.log10(phi2/binwidth02)

# Creating the bin:
logL2=[]
for i in range(0,len(binr2)-1):
    logL2.append((binwidth02/2)+binr2[i])
logL2=np.array(logL2)

######################## Calculating Error: Sqrt(N)#############################

PhiErr1=np.array(1/np.sqrt(count1))
PhiErr2=1/np.sqrt(count2)

######################## 3DSHT Halpha Luminosity Function #############################
plt.figure(figsize = (10,10))
x1, x2 = 40.5,43.5   #  plot x range
y1, y2 = min(logPhi1)-1.5, max(logPhi1)+1 # plot y range

plt.plot(logL1,logPhi1,'r*',ms=12,label='1.0 < z < 1.5')
plt.plot(logL2,logPhi2,'k*',ms=12,label='1.5 < z < 2.3')
plt.errorbar(logL1, logPhi1, yerr=PhiErr1,fmt='r.')
plt.errorbar(logL2, logPhi2, yerr=PhiErr2,fmt='k.')

#plt.axis([x1,x2,y1,y2])
plt.errorbar(x1, y1, xerr=10, yerr=10)
plt.xlabel('$Log[L(erg/s )]$')
plt.ylabel('Log[$\Phi (Mpc^{-3})] dlogL^{-1}$')
plt.title('3D HST: [OIII] Luminosity Function')
plt.legend(loc='upper right',prop={'size':14},frameon=True)

#plt.savefig('3DHST_OIIILumFunc.png')
plt.show()

```
